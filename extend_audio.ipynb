```python
import os
import random
import torch
import torchaudio
import torchaudio.functional as F
```

```python
prefix = "/home/isol/work/bc_resnet_re9ulus/test_audio/1m/positive"
audio_list = sorted(os.listdir("/home/isol/work/bc_resnet_re9ulus/test_audio/1m/positive"))
audio_path_list = [os.path.join(prefix, audio) for audio in audio_list]

noise_path = "/home/isol/work/bc_resnet_re9ulus/test_audio/noise.wav"
noise, noise_sr = torchaudio.load(noise_path)
noise = noise.mean(dim=0, keepdim=True)

for i, audio_path in enumerate(audio_path_list):
    waveform, sr = torchaudio.load(audio_path)

    # 랜덤 소음 가져오기
    front_noise_start_index = random.randint(0, len(noise[0])-8000)
    back_noise_start_index = random.randint(0, len(noise[0])-8000)
    front_noise = noise[:,front_noise_start_index:front_noise_start_index+8000]
    back_noise = noise[:,back_noise_start_index:back_noise_start_index+8000]

    combined = torch.cat([front_noise, waveform, back_noise], dim=1)

    save_path = os.path.join("/home/isol/work/bc_resnet_re9ulus/test_audio/1m/positive_extend", f"positive_{i+1:05d}.wav")
    torchaudio.save(save_path, combined, sr)
```
